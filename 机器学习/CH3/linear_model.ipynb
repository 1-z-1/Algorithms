# 基本形式
线性模型有以下基本形式：
$f(x) = w_1x_1+w_2x_2+\dots+w_dx_d+b$

可以用向量形式写成：
$f(x) = w^Tx+b$

在python中实现一个线性模型

```python
# 导入一些需要的库
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split # 这个库是用来做交叉验证等等操作的，该函数用来切分数据集为测试集和验证集
from sklearn.metrics import mean_squared_error # 这个库有各种各样的评价指标
```

```python
np.random.seed(1)
X = 2*np.random.rand(100,1)
Y = 4 + (-3) * X + np.random.randn(100, 1)
```

```python
# 切分数据
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
```

```python
# 开始训练
lin_reg = LinearRegression()
lin_reg.fit(X_train, Y_train)
```

```python
print("Intercept:", lin_reg.intercept_)
print("Coefficient:", lin_reg.coef_)
```

```python
Y_pred = lin_reg.predict(X_test)

# 计算均方误差
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error:", mse)

# 可视化训练集和拟合直线
plt.scatter(X_train, Y_train, color='blue')
plt.plot(X_train, lin_reg.predict(X_train), color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```
